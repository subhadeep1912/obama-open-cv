```python
img = cv2.imread(obama_img, cv2.IMREAD_COLOR)
plt.imshow(img)
print(img)
```

```python
act_img =img[:, :, ::-1]
plt.imshow(act_img)
```

```python
cropped_img = act_img[0:350, 350:600]
cropped_img
```

```python
plt.imshow(cropped_img)
```

```python
act_img_2x = cv2.resize(cropped_img, None, fx = 2, fy = 2)
plt.imshow(act_img_2x)
```

```python
meme = cv2.resize(act_img_2x, dsize = (700, 300), interpolation=cv2.INTER_AREA)
plt.imshow(meme)
```

```python
meme_copy = meme.copy()
cv2.circle(meme_copy, (500, 120), 50, (0, 0, 255),thickness=5, lineType = cv2.LINE_AA)
plt.imshow(meme_copy)
```

```python
cv2.rectangle(meme_copy, (200, 200), (500, 250), (0,0,255), thickness=5, lineType = cv2.LINE_AA)
plt.imshow(meme_copy)
```

```python
cv2.putText(meme_copy,"BRUH", (150, 75), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale = 2.3, color=(0,0,255), thickness=2, lineType=cv2.LINE_AA)
plt.axis("off")
plt.imshow(meme_copy)
```

```python
meme_copy.shape
```

```python
matrix = np.ones(meme_copy.shape, dtype="uint8") * 90
darker = cv2.subtract(meme_copy, matrix)
bright = cv2.add(meme_copy, matrix)
plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.axis("off")
plt.imshow(darker)
plt.subplot(122)
plt.axis("off")
plt.imshow(bright)
```
